Punto 1. a) realice la suma de las im√°genes

```python
##Primero se importan las librerias que permiten hacer las operaciones matematicas y graficar dichos resultados
import matplotlib.pyplot as plt
import cv2

##se crea una figura para mostrar las dos formas de realizar la suma
fig = plt.figure(figsize=(10,7))

#Luego se cargan las 4 imagenes de la tareas
imagen1 = cv2.imread('waste1.jpg')
imagen2 = cv2.imread('waste2.jpg')
imagen3 = cv2.imread('waste3.jpg')
imagen4 = cv2.imread('waste4.jpg')

#Antes de aplicar las operaciones se ajustan todas las imagenes con las
#mismas dimensiones para que sea posible sumarlas (interpolaci√≥n)
ancho = 680;
alto = 401;
dim = (ancho,alto)
imagen1 = cv2.resize(imagen1,dim, interpolation = cv2.INTER_AREA)
imagen2 = cv2.resize(imagen2,dim, interpolation = cv2.INTER_AREA)
imagen3 = cv2.resize(imagen3,dim, interpolation = cv2.INTER_AREA)
imagen4 = cv2.resize(imagen4,dim, interpolation = cv2.INTER_AREA)

#se designa la posici√≥n de la primera imagen 
fig.add_subplot(1, 2, 1)

#se realiza la suma de las 4 im√°genes usando el operador "+",el cual
#no toma en cuenta el desbordamiento del numero de bits cuando se 
#sobrepasan los 255
sum = imagen1+imagen2+imagen3+imagen4
plt.imshow(sum)
plt.axis('on')
plt.title('suma con operador')

cv2.destroyAllWindows()


#luego se realiza el mismo procedimiento pero usando la operaci√≥n cv2.add
#pues esta operaci√≥n si cualquier suma es superior a 255 lo lleva a tal valor
#para evitar saturaciones.
sum1_2 = cv2.add(imagen1,imagen2) #se suman de a 2 elementos porque el cv2.add recibe m√°ximo 3 entradas
sum3_4 = cv2.add(imagen3,imagen4)
sumtotal = cv2.add(sum1_2,sum3_4)

#se designa la posici√≥n de la segunda imagen 
fig.add_subplot(1, 2, 2)

plt.imshow(sumtotal)
plt.title('suma con cv2.add')
cv2.destroyAllWindows()
```

1.b)realizar la resta de las imagenes

```python
fig = plt.figure(figsize=(10,7))

#se designa la posici√≥n de la primera imagen 
fig.add_subplot(1, 2, 1)

#se realiza la resta de las 4 im√°genes usando el operador "-",el cual
#no toma en cuenta el desbordamiento del numero de bits cuando se 
#sobrepasan los 255
resta = imagen1-imagen2-imagen3-imagen4
plt.imshow(resta)
plt.axis('on')
plt.title('resta con operador')

cv2.destroyAllWindows()

#luego se realiza el mismo procedimiento pero usando la operaci√≥n cv2.subtract
#pues esta operaci√≥n si cualquier resta es superior a 255 lo lleva a tal valor
#para evitar saturaciones.
rest1_2 = cv2.subtract(imagen1,imagen2) #se restan de a 2 elementos porque el cv2.subtract recibe m√°ximo 3 entradas
rest3_4 = cv2.subtract(imagen3,imagen4)
resttotal = cv2.subtract(rest1_2,rest3_4)

#se designa la posici√≥n de la segunda imagen 
fig.add_subplot(1, 2, 2)

plt.imshow(resttotal)
plt.title('resta (1-2)-(3-4) con cv2.subtract')
cv2.destroyAllWindows()
```

1.d)Realizar la multiplicaci√≥n de las im√°genes

```python
fig = plt.figure(figsize=(10,7))

#se designa la posici√≥n de la primera imagen 
fig.add_subplot(1, 2, 1)

#se realiza la multiplicacion de las 4 im√°genes usando el operador "*",el cual
#no toma en cuenta el desbordamiento del numero de bits cuando se 
#sobrepasan los 255
multiplicacion = imagen1*imagen2*imagen3*imagen4
plt.imshow(multiplicacion)
plt.axis('on')
plt.title('multiplicacion con operador')

cv2.destroyAllWindows()

#luego se realiza el mismo procedimiento pero usando la operaci√≥n cv2.multiply
#pues esta operaci√≥n si cualquier multiplicacion es superior a 255 lo lleva a tal valor
#para evitar saturaciones.
mult1_2 = cv2.multiply(imagen1,imagen2) #se multiplican de a 2 elementos porque el cv2.add reciba m√°ximo 3 entradas
mult3_4 = cv2.multiply(imagen3,imagen4)
multtotal = cv2.multiply(mult1_2,mult3_4)

#se designa la posici√≥n de la segunda imagen 
fig.add_subplot(1, 2, 2)

plt.imshow(multtotal)
plt.title('multiplicacion (1*2)*(3*4) con cv2.multiply')
cv2.destroyAllWindows()
```

1.c)Realizar la division de las im√°genes

```python
fig = plt.figure(figsize=(10,7))

#se designa la posici√≥n de la primera imagen 
fig.add_subplot(1, 2, 1)

#se realiza la division de las 4 im√°genes usando el operador "*",el cual
#no toma en cuenta el desbordamiento del numero de bits cuando se 
#sobrepasan los 255
division = imagen1/imagen2/imagen3/imagen4
plt.imshow(division)
plt.axis('on')
plt.title('division con operador')

cv2.destroyAllWindows()

#luego se realiza el mismo procedimiento pero usando la operaci√≥n cv2.multiply
#pues esta operaci√≥n si cualquier multiplicacion es superior a 255 lo lleva a tal valor
#para evitar saturaciones.
div1_2 = cv2.divide(imagen1,imagen2) #se dividen de a 2 elementos porque el cv2.add reciba m√°ximo 3 entradas
div3_4 = cv2.divide(imagen3,imagen4)
divtotal = cv2.divide(div1_2,div3_4)

#se designa la posici√≥n de la segunda imagen 
fig.add_subplot(1, 2, 2)

plt.imshow(divtotal)
plt.title('division (1/2)/(3/4) con cv2.divide')
cv2.destroyAllWindows()
```

1.e) Realice las operaciones bitwise

```python
##Bitwise not
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2, 2, 1)
not1 = ~imagen1
plt.imshow(not1)
fig.add_subplot(2, 2, 2)
not2 = ~imagen2
plt.imshow(not2)
fig.add_subplot(2, 2, 3)
not3 = ~imagen3
plt.imshow(not3)
fig.add_subplot(2, 2, 4)
not4 = ~imagen4
plt.imshow(not4)


cv2.destroyAllWindows()
```

```python
#bitwise and
fig = plt.figure(figsize=(10,7))
fig.add_subplot(1, 2, 1)
imag1_2 = imagen1 & imagen2
plt.imshow(imag1_2)
plt.title('imagen 1 & 2')
fig.add_subplot(1, 2, 2)
imag3_4 = imagen3 & imagen4
plt.imshow(imag3_4)
plt.title('imagen 3 & 4')
cv2.destroyAllWindows()
```

```python
##bitwise or
fig = plt.figure(figsize=(10,7))
fig.add_subplot(1, 2, 1)
imag1or2 = imagen1 | imagen2
plt.imshow(imag1or2)
plt.title('imagen 1 | 2')
fig.add_subplot(1, 2, 2)
imag3or4 = imagen3 | imagen4
plt.imshow(imag3or4)
plt.title('imagen 3 | 4')
cv2.destroyAllWindows()
```

```python
##bitwise xor
fig = plt.figure(figsize=(10,7))
fig.add_subplot(1, 2, 1)
imag1xor2 = imagen1 ^ imagen2
plt.imshow(imag1xor2)
plt.title('imagen 1 ^ 2')
fig.add_subplot(1, 2, 2)
imag3xor4 = imagen3 ^ imagen4
plt.imshow(imag3xor4)
plt.title('imagen 3 ^ 4')
cv2.destroyAllWindows()
```
